```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)
fig, ax = plt.subplots()
ax.set_axis_off()
Z = np.sin(X)
img = ax.imshow(Z, cmap='plasma', animated=True)
def update(frame):
    Z = np.sin(X + frame/5.0) + np.sin(Y + frame/10.0)
    img.set_array(Z)
    return [img]
ani = animation.FuncAnimation(fig, update, frames=100, interval=50, blit=True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 800)
y = np.linspace(0, 10, 800)
X, Y = np.meshgrid(x, y)

Z = np.sin(X**2 + Y**2) + np.cos(X * Y)

plt.figure(figsize=(8, 8))
plt.imshow(Z, cmap='inferno', interpolation='bilinear')
plt.axis('off')
plt.title("🎨 Abstract Sinusoidal Art")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = 15  
magic_square = np.zeros((n, n), dtype=int)
i, j = 0, n // 2
for num in range(1, n*n + 1):
    magic_square[i, j] = num
    newi, newj = (i - 1) % n, (j + 1) % n
    if magic_square[newi, newj]:
        i += 1
    else:
        i, j = newi, newj
plt.imshow(magic_square, cmap='coolwarm')
plt.title(f"Magic Square {n}x{n}")
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
days = 100
dates = pd.date_range("2025-01-01", periods=days)
price = 100 + np.cumsum(np.random.randn(days) * 2)  
df = pd.DataFrame({'Date': dates, 'Price': price})
df.set_index('Date', inplace=True)
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Price'], label="Simulated Stock Price")
plt.title("📈 Simulated Stock Market Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Title': [
        'Inception', 'The Matrix', 'Interstellar', 'The Godfather', 'Parasite',
        'Avengers: Endgame', 'The Dark Knight', 'Fight Club', 'Pulp Fiction', 'Forrest Gump'
    ],
    'Year': [2010, 1999, 2014, 1972, 2019, 2019, 2008, 1999, 1994, 1994],
    'Rating': [8.8, 8.7, 8.6, 9.2, 8.6, 8.4, 9.0, 8.8, 8.9, 8.8]
}
df = pd.DataFrame(data)
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
df_sorted = df.sort_values(by='Rating', ascending=False)
plt.barh(df_sorted['Title'], df_sorted['Rating'], color='teal')
plt.xlabel('Rating')
plt.title('Top 10 Movies by Rating')
plt.gca().invert_yaxis()  # Highest rating on top
plt.subplot(1, 2, 2)
yearly_avg = df.groupby('Year')['Rating'].mean()
plt.plot(yearly_avg.index, yearly_avg.values, marker='o', linestyle='-', color='darkorange')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Movie Rating Over Years')
plt.grid(True)
plt.tight_layout()
plt.show()
```
